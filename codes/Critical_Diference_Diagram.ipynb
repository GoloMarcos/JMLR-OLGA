```
# gae & 4.375\\
# deepwalk & 3.375\\
# node2vec & 4.1875\\
# OC-GCN & 5.1875\\
# OC-GAT & 4.125\\
# OC-GraphSAGE & 5.0625\\
# OLGA & 1.6875\\

names = ['OCGCN', 'OCSAGE', 'OCGAT', 'Node2Vec', 'GAE', 'Deepwalk', 'OLGA']

avranks =  [5.1875, 5.0625, 4.125, 4.1875, 4.375, 3.375, 1.6875]

cd = Orange.evaluation.scoring.compute_CD(avranks, 16) # tested on 16 scenarios

Orange.evaluation.scoring.graph_ranks(avranks, names, cd=cd,filename="olga_nem.pdf",  width=10, textspace=1.3)
print(cd)
```
