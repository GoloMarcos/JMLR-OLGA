```python
from pathlib import Path
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt

def plot_2d_baselines(dataset, feature):
  pt = '/content/drive/MyDrive/USP/Doctorate/Research/Articles/OLGA: One cLass Graph Autoencoder/datasets/graphs/'

  basepath = Path(pt)
  path = pt + dataset + '/k=1/' + dataset + '_k=1_fold=0.gpickle'

  graph = nx.read_gpickle(path)

  embs = []
  colors = []
  for node in graph.nodes():
    embs.append(graph.nodes[node][feature])
    if graph.nodes[node]['label'] == 1:
      colors.append('#0074ff')
    else:
      colors.append('#00ff5c')

  embs = np.array(embs)

  fig = plt.figure(figsize=(8, 8))
  ax = fig.add_subplot()
  ax.clear()
  ax.scatter(embs[:, 0], embs[:, 1], s=100, c=colors, cmap="hsv")
  plt.show()
```

```python
for dataset in ['relevant_reviews', 'food', 'TUANDROMD']:
  print(dataset)
  for feature in ['features_gae_2','features_node2vec_2','features_deepwalk_2']:
    print(feature)
    plot_2d_baselines(dataset, feature)
```
